```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori

dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])

rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)
results = list(rules)

```

```python
items=[[]]
for i in range(len(results)):
    items.append([str(', '.join([x for x in results[i][0]])),float(results[i][2][0][2])*100] ) 
```

```python
final_items = sorted(items[1:],key=lambda x: x[1], reverse=True)

print('The Association rules are updated in Association_rule.csv file\n')
for i,item in enumerate(final_items):
    print('{}. {} are bought together | Confidence = {}'.format(i+1,item[0],item[1]))
    
    
df = pd.DataFrame(final_items)
df.columns = ['Items bought together','Confidence']
df.head(20)
df.to_csv('Association_rule.csv')
```
